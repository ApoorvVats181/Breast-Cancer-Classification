# Import libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Import dataset

```python
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
```

```python
print(cancer.keys())
```

```python
print(cancer['target_names'])
```

```python
print(cancer['feature_names'])
```

```python
print(cancer['data'].shape)
```

```python
df = pd.DataFrame(np.c_[cancer['data'], cancer['target']], columns = np.append(cancer['feature_names'], ['target']))
```

```python
df.head()
```

```python
df.tail()
```

# Visualizing data

```python
# correlation between different variables
sns.pairplot(df, hue = 'target', vars = ['mean radius', 'mean texture', 'mean area', 'mean perimeter', 'mean smoothness'] )
```

```python
sns.countplot(df['target'])
```

```python
sns.scatterplot(x = 'mean area', y = 'mean smoothness', hue = 'target', data = df)
```

```python
plt.figure(figsize = (20,10))
sns.heatmap(df.corr(), annot = True)
```

# Training model

```python
x = df.drop(['target'],axis=1)
y = df['target']
```

# Splitting data

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 5)
```

```python
from sklearn.svm import SVC
model = SVC()
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot = True)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print('Accuracy score : ',accuracy_score(y_test,y_pred)*100,'%')
```

# IMPROVING THE MODEL BY NORMALIZING DATA

# Normalize training data

```python
min_train = x_train.min()
range_train = (x_train - min_train).max()
x_train_scaled = (x_train - min_train)/range_train
```

```python
# before scaling
sns.scatterplot(x = x_train['mean area'], y = x_train['mean smoothness'], hue = y_train)
```

```python
# after scaling
sns.scatterplot(x = x_train_scaled['mean area'], y = x_train_scaled['mean smoothness'], hue = y_train)
```

# Normalize testing data

```python
min_test = x_test.min()
range_test = (x_test - min_test).max()
x_test_scaled = (x_test - min_test)/range_test
```

```python
model.fit(x_train_scaled, y_train)
```

```python
y_predict = model.predict(x_test_scaled)
```

```python
cm = confusion_matrix(y_test, y_predict)
sns.heatmap(cm, annot = True)
```

```python
print('Accuracy score(After Normalizing) : ',accuracy_score(y_test,y_predict)*100,'%')
```

```python
print(classification_report(y_test, y_predict))
```

# IMPROVING THE MODEL BY ADJUSTING PARAMETERS

# Apply Grid Search Method

```python
from sklearn.model_selection import GridSearchCV
```

```python
param = { 
          'C' : [0.1, 1, 10, 100],
          'gamma' : [1, 0.1, 0.01, 0.001],
          'kernel' : ['rbf']
        }
```

```python
grid_search = GridSearchCV(
                           estimator = model,
                           param_grid = param,
                           scoring = 'accuracy',
                           cv = 10,
                           n_jobs = -1,
                           #refit = True,
                           #verbose = 4
                          )
```

```python
grid_search.fit(x_train_scaled, y_train)
```

```python
grid_search.best_params_
```

```python
grid_pred = grid_search.predict(x_test_scaled)
```

```python
cm = confusion_matrix(y_test, grid_pred)
sns.heatmap(cm, annot = True)
```

```python
acc = accuracy_score(y_test, grid_pred)
prec = precision_score(y_test, grid_pred)
rec = recall_score(y_test, grid_pred)
f1 = f1_score(y_test, grid_pred)
```

```python
print('Accuracy score(After Normalizing): ',acc*100,'%')
print('Precision: ',prec)
print('Recall: ',rec)
print('F1 score: ',f1)
```
